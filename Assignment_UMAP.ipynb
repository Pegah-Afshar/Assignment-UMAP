```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from sklearn.datasets import fetch_openml
```

1.Load the dataset and conduct any necessary preprocessing, such as normalizing the data.

```python
mnist = fetch_openml('Fashion-MNIST', version=1, cache=True)
```

```python
indice=np.random.choice(70000,10000)
X=mnist.data[indice]/255
y=mnist.target[indice]
print(X.shape,y.shape)
```

2. Apply UMAP to the data.

```python
! pip install umap-learn
```

```python
import umap
```

```python
umap_result=umap.UMAP(n_components=2,random_state=1).fit_transform(X)
```

3. Using the two-dimensional UMAP representation, draw a graph of the data by coloring and labeling the data points as you did in the checkpoint.

```python
plt.figure(figsize=(10,5))
colours = ["r","b","g","c","m","y","k","r","burlywood","chartreuse"]
for i in range(umap_result.shape[0]):
    plt.text(umap_result[i, 0], umap_result[i, 1], y[i],
             color=colours[int(y[i])],
             fontdict={'weight': 'bold', 'size': 50}
        )

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(umap_result[:,0],umap_result[:,1])
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.show()
```

5. Now, play with the different hyperparameter values of the UMAP, and apply UMAP for each of them. Which combination is the best in terms of the two-dimensional representation clarity?

```python
umap_result=umap.UMAP(n_components=2,n_neighbors=5,min_dist=0.3,metric='correlation',random_state=1).fit_transform(X)
plt.figure(figsize=(10,5))
plt.scatter(umap_result[:,0],umap_result[:,1])

plt.axis('off')
plt.show()
```

```python
umap_result=umap.UMAP(n_components=2,n_neighbors=10,min_dist=0.3,metric='correlation',random_state=1).fit_transform(X)
plt.figure(figsize=(10,5))
plt.scatter(umap_result[:,0],umap_result[:,1])

plt.axis('off')
plt.show()
```
